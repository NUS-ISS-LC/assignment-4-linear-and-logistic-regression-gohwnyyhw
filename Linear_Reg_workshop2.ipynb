# Linear Regression Example 

Import all the libraries needed and load the dataset

```python
# import everything we need
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('OmniPower.csv')
```

Construct a dataframe for the data (independent variable) and target (dependent variable)

```python
input_df = df[['Price', 'Promotion']]
sales = df['Sales'] 
# Display the first five rows of the data
input_df.head()
```

Prepare the model training data based on the independent and dependent variables

```python
# prepare the training data
from sklearn import linear_model
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input_df,sales, test_size=0.2, random_state = 42)
```

Create a Linear Regression Model and train the model

```python
# instantiate a linear regression model
# train the model to fit the training data, finding the coef and intercept

# import model
from sklearn.linear_model import LinearRegression

#instantiate
linReg = LinearRegression()

# fit out linear model to the train set data
linReg.fit(x_train, y_train)
```

After training the model, predict the sales using the test data

```python
from sklearn.metrics import r2_score

y_pred = linReg.predict(x_test) 
print(y_pred)
```

Model Validation

```python
# Diaplay the coefficient, intercept and r_square value of our model
from sklearn.metrics import r2_score

print("Coefficient: ",linReg.coef_)
print("Intercept: ", linReg.intercept_)
print('r2_score: ',r2_score(y_test, y_pred)) 
```

Visualise to see what the model looks like

```python
# Plot the ideal regression line
# Find the min and max values to define the range of the line
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
reg_line = np.linspace(min_val, max_val, 100)

# Plot the regression line in blue
plt.plot(reg_line, reg_line, color='blue')
# Plot the actual value in red
plt.scatter(y_test, y_pred, color='red', label='Actual vs. Predicted Values', alpha=0.6)
plt.show()
```
